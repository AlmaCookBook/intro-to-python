# An introduction to programming with Python

## Outline for self-study

Work through each of these notebooks in order, checking the solutions  in the solutions directory as you go. If you have any questions you can contact scsoftware.icr.ac.uk or try the python-training slack group in Scientific Computing Services. It is intended that the material might take roughly 2 days. The first day is mostly acquiring concepts and the second day is mostly practice.

## Outline for in-person 2 day course

The in-person (or online) courses take the form of a guided workshop. There is an element of micro-flipped-learning: we go through each module first individually / talking to colleagues / asking for guidance, before coming back together as a group and looking for the solution by consensus.
